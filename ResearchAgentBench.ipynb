# ScientistBench Evaluation Framework v1

## Task Categories

1. Literature Summarization
2. Literature Synthesis
3. Research Gap Discovery
4. Hypothesis Generation

## Evaluation Dimensions

### Accuracy (0-5)

Is the answer factually correct?

### Completeness (0-5)

Does the answer cover the important points?

### Reasoning Quality (0-5)

Does the answer demonstrate logical reasoning?

### Evidence Support (0-5)

Is the answer supported by the provided context?

### Novelty (0-5)

(Only for Gap Discovery and Hypothesis Generation)

Does the answer provide meaningful new insights?

## Final Score

Final Score =
Accuracy +
Completeness +
Reasoning +
Evidence Support +
Novelty

Maximum Score = 25

```python
client = arxiv.Client()

search = arxiv.Search(
    query="""
    agentic ai OR
    scientific discovery OR
    ai research assistant
    """,
    max_results=50,
    sort_by=arxiv.SortCriterion.Relevance
)

papers = []

for result in client.results(search):

    papers.append({
        "title": result.title,
        "published": result.published,
        "summary": result.summary,
        "pdf_url": result.pdf_url
    })

papers_df = pd.DataFrame(papers)

print("Total Papers:", len(papers_df))
papers_df.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt

categories = list(category_results.columns)

N = len(categories)

angles = np.linspace(
    0,
    2*np.pi,
    N,
    endpoint=False
).tolist()

angles += angles[:1]

fig = plt.figure(figsize=(8,8))
ax = plt.subplot(111, polar=True)

for model in category_results.index:
    values = category_results.loc[model].tolist()
    values += values[:1]

    ax.plot(
        angles,
        values,
        linewidth=2,
        label=model
    )

    ax.fill(
        angles,
        values,
        alpha=0.05
    )

ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)

plt.title(
    "ResearchAgentBench Category Radar Comparison"
)

plt.legend(
    loc="upper right",
    bbox_to_anchor=(1.3,1.1)
)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

means = leaderboard["mean"]
stds = leaderboard["std"]
n = leaderboard["count"]

ci = 1.96 * (stds / np.sqrt(n))

plt.figure(figsize=(8,5))

plt.bar(
    means.index,
    means.values,
    yerr=ci.values,
    capsize=8
)

plt.title(
    "95% Confidence Intervals of Model Scores"
)

plt.ylabel("Average Score")

plt.show()
```

```python
from scipy.stats import friedmanchisquare

chatgpt = results_df[
    results_df["model"]=="ChatGPT"
]["total_score"]

claude = results_df[
    results_df["model"]=="Claude"
]["total_score"]

deepseek = results_df[
    results_df["model"]=="DeepSeek"
]["total_score"]

gemini = results_df[
    results_df["model"]=="Gemini"
]["total_score"]

stat, p = friedmanchisquare(
    chatgpt,
    claude,
    deepseek,
    gemini
)

print("Friedman Statistic:", stat)
print("p-value:", p)
```
